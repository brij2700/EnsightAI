## Create Daily Values based on timeslots for all users in a block


```python
import pandas as pd
for block_no in range(28,112):
    
    # Read the first CSV file
    df1 = pd.read_csv(f"..\\Dataset\\archive\daily_dataset.csv")
    
    # Read the second CSV file
    df2 = pd.read_csv(f"..\\Dataset\\archive\\hhblock_dataset\\hhblock_dataset\\block_{block_no}.csv")
    
    # Get unique LCLids from block_22.csv
    unique_lids = df2['LCLid'].unique()
    
    # List to store individual result dataframes
    result_dfs = []
    
    # Loop through each LCLid
    for lid in unique_lids:
        # Filter data for the current LCLid in both dataframes
        df1_filtered = df1[df1['LCLid'] == lid]
        df2_filtered = df2[df2['LCLid'] == lid]
        
        # Merge the two dataframes on 'LCLid' and 'day'
        merged_df = pd.merge(df1_filtered, df2_filtered, on=['LCLid', 'day'])
        
        # Define a function to calculate morning, afternoon, evening, and night
        def categorize_hours(row):
            morning = row['hh_12':'hh_24'].sum()
            afternoon = row['hh_25':'hh_32'].sum()
            evening = row['hh_33':'hh_40'].sum()
            night = row.drop(['LCLid', 'day', 'energy_median', 'energy_mean', 'energy_max', 'energy_count', 'energy_std', 'energy_sum', 'energy_min']).sum() - morning - afternoon - evening
            return pd.Series([morning, afternoon, evening, night], index=['morning', 'afternoon', 'evening', 'night'])
    
        # Apply the function to each row
        new_df = merged_df.apply(categorize_hours, axis=1)
    
        # Concatenate the original columns with the categorized hours
        result_df = pd.concat([merged_df[['LCLid', 'day']], new_df], axis=1)
        
        # Append the result dataframe to the list
        result_dfs.append(result_df)
    
    # Concatenate all the result dataframes
    final_result = pd.concat(result_dfs)
    final_result['total_consumption'] = final_result['morning'] + final_result['afternoon'] + final_result['evening'] + final_result['night']

    final_result.to_csv(f"..\\Dataset\\Created\\Daily\\block{block_no}_all_lids_dailyhours.csv", index=False)
    print("Done:",block_no)
    # Write the final result to a new CSV file

```
