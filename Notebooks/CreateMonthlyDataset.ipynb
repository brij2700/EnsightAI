```python
import pandas as pd
for block_no in range(1,112):
    # Read the final result CSV file with total consumption
    final_result_df = pd.read_csv(f"..\\Dataset\\Created\\Daily\\block{block_no}_all_lids_dailyhours.csv")
    
    # Convert 'day' column to datetime datatype
    final_result_df['day'] = pd.to_datetime(final_result_df['day'])
    
    # Extract month and year from the 'day' column
    final_result_df['month'] = final_result_df['day'].dt.month
    final_result_df['year'] = final_result_df['day'].dt.year
    
    # Group by month, year, and LCLid, and sum up the energy consumption for each time period
    consolidated_df = final_result_df.groupby(['LCLid', 'year', 'month']).agg({
        'morning': 'sum',
        'afternoon': 'sum',
        'evening': 'sum',
        'night': 'sum',
        'total_consumption': 'sum'
    }).reset_index()
    
    # Write the consolidated result to a new CSV file
    consolidated_df.to_csv(f"..\\Dataset\\Created\\Monthly\\block{block_no}_consolidated_usage_by_month.csv", index=False)
    print("Done",block_no)
```
